## Introduction
### Background Information
Stories, screenwriting, casting, shooting, and editing are some of the essential factors of filmmaking. That is, creating a movie takes people from various sectors, including story writers, directors, investors, and cast members. Since movie productions consist of multiple components, it is usually a misnomer to say that one single factor single-handedly 'carried', or led, the movie to success. Hence, we almost never see a story writer, director, or cast member occupying an overwhelming share in the whole film industry. However, producers still have preferences when it comes to casting actors and actresses. It could be that they are well suited to the characters themselves, but producers also wish to recruit celebrities who effortlessly attract a mainstream audience with their looks and natural talent.<br>

Also, certain genres have been consistently popular. According to an [analysis from 'Stephen Follows'](https://stephenfollows.com/genre-trends-global-film-production/), 'Drama' takes up over half of the genre of global film production from 1988 to 2017, followed by 'Comedy' with over 25%. While 'Drama' and 'Comedy' movies are aimed to draw attraction from a wide spectrum of the population, many of the other movie genres tend to focus on frantic fans. It is also surprising to realize that popular movies aren't necessarily the most profitable ones.<br>

As such, there are several intriguing facts regarding the film industry that is worth analyzing. With this in mind, we thought it would be interesting to examine how these genres compete for popularity, better audience appraisal, and how profitable each of them is.

### Key terms
- Genre: a style of film that involves a particular set of characteristics; in the dataset that we retrieved, films are categorized as at least one, up to four genres
- Profit: money that is earned in the production of the film after paying the costs of production; here, the value will be calculated as `Worldwide Gross Income - Budget`
- Popularity: [TMDB measure](https://developers.themoviedb.org/3/getting-started/popularity) that takes into account 'number of votes for the day', 'number of views for the day', 'release date', 'number of total votes', and 'previous days score'
- Rating (Vote Average): audience's vote on how much they like a film ranging from 0 to 10

```python
# loading libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import re
from sklearn.linear_model import LinearRegression
import seaborn as sns
# loading data
raw_movie_data = pd.read_csv("dataset/tmdb_5000_movies.csv")
raw_cast_data = pd.read_csv("dataset/tmdb_5000_credits.csv")
# the cast data would add relevant information about the movie cast and crew, while "movie" data has the rest of the attributes 
```

```python
# the "genres" column includes up to 6 genres one movie can be affiliated with, so we want to create 6 columns corresponding to genres of each movie
new = combined_data["genres"].str.split(",", n=11, expand = True) 
new.columns = ["1", "genre_1", "2", "genre_2", "3", "genre_3", "4", "genre_4", "5", "genre_5", "6", "genre_6"] 
new.head(3)
```

```python
# we want to check how many movies actually might have all 6 genres they fall into
#since only 100 movies have a 6th genre, there is no pressing need to include it in our future analysis
new["genre_6"].isna().sum()
```

```python
# since the data lists a movie's genres based on importance in descending order, we will only keep information about the first 5 genres
new = new[ [ 'genre_1', 'genre_2', 'genre_3', 'genre_4', 'genre_5'] ] 

# now we can clean the string values within genres from unnecessary characters
new["genre_1"] = new["genre_1"].str.replace('name', '').str.lstrip(': "').str.rstrip('"}]').str.rstrip('"}')
new["genre_2"] = new["genre_2"].str.replace('name', '').str.lstrip(': "').str.rstrip('"}]').str.rstrip('"}')
new["genre_3"] = new["genre_3"].str.replace('name', '').str.lstrip(': "').str.rstrip('"}]').str.rstrip('"}')
new["genre_4"] = new["genre_4"].str.replace('name', '').str.lstrip(': "').str.rstrip('"}]').str.rstrip('"}')
new["genre_5"] = new["genre_5"].str.replace('name', '').str.lstrip(': "').str.rstrip('"}]').str.rstrip('"}')
new["genre_5"] = new["genre_5"].str.replace('name', '').str.lstrip(': "').str.rstrip('"}]')

#we create a new dataframe that has the genres columns and add modie id to it for future merging
movie_id = combined_data["id"]
new = new.join(movie_id)
new.head(3)
```

```python
# creating new function to convert the movie budgets in international currencies into USD
# the exchange rates were taken from Morningstar in Google on Sep 2020
def convert_rates(money):
    if 'EUR' in money:
        return int(''.join(x for x in money if x.isdigit())) * 1.17
    elif 'GBP' in money:
        return int(''.join(x for x in money if x.isdigit())) * 1.29
    elif 'CNY' in money:
        return int(''.join(x for x in money if x.isdigit())) * 0.15
    elif 'CAD' in money:
        return int(''.join(x for x in money if x.isdigit())) * 0.75
    elif 'DEM' in money:
        return int(''.join(x for x in money if x.isdigit())) * 0.6
    elif 'THB' in money:
        return int(''.join(x for x in money if x.isdigit())) * 0.032
    elif 'FRF' in money:
        return int(''.join(x for x in money if x.isdigit())) * 0.18
    elif 'AUD' in money:
        return int(''.join(x for x in money if x.isdigit())) * 0.71
    elif 'HUF' in money:
        return int(''.join(x for x in money if x.isdigit())) * 0.0033
    elif 'INR' in money:
        return int(''.join(x for x in money if x.isdigit())) * 0.014
    elif 'DKK' in money:
        return int(''.join(x for x in money if x.isdigit())) * 0.16
    elif 'NOK' in money:
        return int(''.join(x for x in money if x.isdigit())) * 0.11
    elif 'BRL' in money:
        return int(''.join(x for x in money if x.isdigit())) * 0.18
    else:
        return int(''.join(x for x in money if x.isdigit()))
```

```python
# using the function to convert the movies' budgets into dollars where needed
final["budget"] = final["budget"].apply(convert_rates)
```

Mapping each genre into number for the linear regression: the numbers will be ascending representative of the genre size (lowest is the smallest genre, highest is the biggest genre).

```python
genres_18_map = {"Foreign": 1, "Western": 2, "War": 3, "Music": 4, "History": 5, "Animation": 6, "Mystery": 7, "Fantasy": 8, "Horror": 9, "Family": 10, "Science Fiction": 11, "Crime": 12, "Adventure": 13, "Romance": 14, "Action": 15, "Thriller": 16, "Comedy": 17, "Drama": 18 }
```

```python
new_tablez = genres_table.copy()
new_tablez["genre_1"] = new_tablez["genre_1"].map(genres_18_map)
new_tablez.head(3)
```

#### The multivariable regression model with genre and budget as variables and revenue as outcome

```python
multivariable_model = LinearRegression().fit(new_tablez[["genre_1", "budget"]], new_tablez[["usa_gross_income"]] )
multivariable_coef = multivariable_model.coef_
print("Genre coefficient:", multivariable_coef.item(0))
print("Budget coefficient:", multivariable_coef.item(1))
print("r^2 score:", multivariable_model.score(new_tablez[["genre_1", "budget"]], new_tablez[["usa_gross_income"]]))
```

This model shows that increasing by 1 in "genre_1" - going from smallest to the biggest genre - changes the revenue by decreasing it by $1.1 million. That could be explained by the fact that the biggest genres have a big distribution of both successful and unsuccessful movies (some generated more revenue than their budget, some not so much). The model's budget coefficient implies that for the movies in our dataset, increasing budget by 1 dollar would most likely result in a positive 1.04 dollar increase in USA gross income. 

#### Single linear regression of budget and revenue for each genre separately

```python
# function for retrieving name of each genre from the mapped number
def get_val(v, d):
    for key, val in d.items():
        if v == val:
            return key
```

```python
for x in range(2,18):
    a_genre = new_tablez.loc[new_tablez['genre_1'] == x]
    a_multivariable_model = LinearRegression().fit(a_genre[["budget"]], a_genre[["usa_gross_income"]] )
    a_multivariable_coef = a_multivariable_model.coef_
    print(get_val(x, genres_18_map))
    print("Budget coefficient: {:.3f}".format(a_multivariable_coef.item(0)))
    print("r^2 score: {:.3f}".format(a_multivariable_model.score(a_genre[[ "budget"]], a_genre[["usa_gross_income"]])))
```

## Limitations
The outline of potential problems with our chosen dataset:
- Since our dataset includes around 3,600 movies with the majority being popular to a certain degree (predominantly in the US market), it might seem like our data can be a limited representation of the movies in the market; that means that our outcomes and analysis of the genres trends can turn out biased in terms of generating genre trends only from a small number of movies. In other words, the dataset we will use might seem like an inaccurate sample taken from a big population of variables. 
- It is important to note that our dataset only includes movies up to year 2016. Therefore, there is a 4-year gap of movies released between 2016 and 2020, so we would not be able to work with the recent movies and idenfity the genres trending during the recent years. 
- For this project, it is merely impossible to be able to examine all the movies (even with focusing on one country’s market) because of the immense size of movie markets. Even if we hope that our dataset has those movies that has a great influence on the genres popularity during certain years, there is a chance that our data analytics will lead to figuring out erroneous trends. 
- Our analysis of financial performance as well as application of budget and revenu calculation of the entire genre are not precise since we did not account for the time value of money (our time period is too wide and external factors like inflation and economy events impacted the value of money).
- If our team project will potentially be used by someone trying to  predict how a genre of a movie influences its success in the future,  our relatively narrow range of movies can bring harmful consequences for that person. Particularly,  those people could be misinformed and later produce a movie that would not reach high popularity and a large box office. 
- We are heavily relying on the dataset’s accurately distributing which movie belongs to which combination of genres; hence, we have to recognize the potential imprecision of results if the genres were allocated for a certain movie wrongly.
