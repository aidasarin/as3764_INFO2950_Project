### Manipulations with Genres 

Some of the most meaningful points from the exploratory data analysis are presented below. These calculations and visuals are mainly important because they help find the answers to our research questions. First, we will calculate how many movies from the analysis-ready dataset belong to each genre, allowing us to then estimate the biggest encompassing genres (within four 20-year timeframes and the entire period). As a result, we could notice the pattern of rise and fall of different genres over time. 

It's important to emphasize that we base our results of genre popularity on the **supply-demand theory** - if a genre had the highest number of its movies released within a certain period, it means that the directors were particularly leaning towards producing movies of that genre. That's why we assume that the most frequent genre could aslo be one of the most popular ones at the time.

Another metrics for establishing genre trends could be the calculated **average popularity score** across each genre. Even though one genre might not have been the largest over time, a few of the world-renowned movies belonging to it could have made the genre extremely popular for years.

```python
# we want to count how many movies fall under each category of genre

genre1_df = final.loc[:, ~final.columns.isin(['genre_2', 'genre_3', 'genre_4', 'genre_5'])]
genre2_df = final.loc[:, ~final.columns.isin(['genre_1', 'genre_3', 'genre_4', 'genre_5'])]
genre3_df = final.loc[:, ~final.columns.isin(['genre_1', 'genre_2', 'genre_4', 'genre_5'])]
genre4_df = final.loc[:, ~final.columns.isin(['genre_1', 'genre_2', 'genre_3', 'genre_5'])]
genre5_df = final.loc[:, ~final.columns.isin(['genre_1', 'genre_2', 'genre_3', 'genre_4'])]


genre1_df = genre1_df.rename(columns={"genre_1": "genre"})
genre2_df = genre2_df.rename(columns={"genre_2": "genre"})
genre3_df = genre3_df.rename(columns={"genre_3": "genre"})
genre4_df = genre4_df.rename(columns={"genre_4": "genre"})
genre5_df = genre5_df.rename(columns={"genre_5": "genre"})


genre2_df = genre2_df[genre2_df['genre'] != 'None']
genre3_df = genre3_df[genre3_df['genre'] != 'None']
genre4_df = genre4_df[genre4_df['genre'] != 'None']
genre5_df = genre5_df[genre5_df['genre'] != 'None']


frames = [genre1_df, genre2_df, genre3_df, genre4_df, genre5_df]
genre_df = pd.concat(frames)

genre_df = genre_df.sort_values(['genre'])

genre_table = genre_df.groupby('genre').count()
genre_table.id.value_counts
```

### Multivariable Linear Regression Model with Popularity, Vote Average and Budget predictors and Revenue outcome


```python
final.corr().iloc[1:,7]
```

We can see that `popularity`, `vote_count`, and `budget` have a high correlation with `worldwide_gross_income`.

```python
y = final['worldwide_gross_income']
features = final[['popularity', 'vote_average', 'budget']]
```

```python
reg = LinearRegression().fit(features, y)
score = reg.score(features, y)
print("Popularity Coefficient:{}".format(reg.coef_[0]))
print("Vote Average Coefficient:{}".format(reg.coef_[1]))
print("Budget Coefficient:{}".format(reg.coef_[2]))
print("r^2 score: {}".format(score))
```

This multivariable linear regression analysis had shown that by increasing the popularity ranking by 1, movies' revenues tend to change by increasing by 1.685 million dollars. The regression coefficient for the vote average is even higher - change in the vote average by 1 causes a worldwide revenue increase by  22.891 millions USD. The budget coefficient means that every dollar invested into a movie is typically brings an average of a $2.5 return.

### The multivariable regression model with genre and budget as variables and revenue as outcome

```python
multivariable_model = LinearRegression().fit(new_tablez[["genre_1", "budget"]], new_tablez[["usa_gross_income"]] )
multivariable_coef = multivariable_model.coef_
print("Genre coefficient:", multivariable_coef.item(0))
print("Budget coefficient:", multivariable_coef.item(1))
print("r^2 score:", multivariable_model.score(new_tablez[["genre_1", "budget"]], new_tablez[["usa_gross_income"]]))
```

This model shows that increasing by 1 in "genre_1" - going from smallest to the biggest genre - changes the revenue by decreasing it by $1.1 million. That could be explained by the fact that the biggest genres have a big distribution of both successful and unsuccessful movies (some generated more revenue than their budget, some not so much). The model's budget coefficient implies that for the movies in our dataset, increasing budget by 1 dollar would most likely result in a positive 1.04 dollar increase in USA gross income. 

### Single linear regression of budget and revenue for each genre separately

```python
# function for retrieving name of each genre from the mapped number
def get_val(v, d):
    for key, val in d.items():
        if v == val:
            return key
```

Our second analysis showed us that "Family" seems the most profitable genre with the highest budget coefficient equal to 1.5 , meaning that for every dollar spent in the movie budget, there would be a 1.5 dollar increase in the movie's revenue. ## Evaluation of Significance

To apply the results from our data analysis on our final conclusions and interpretations, they first require an evaluation of their statistical significance - the degree of unlikeliness of these results occurring randomly. 

### Budget and Popularity Correlation Statistical Significance

It's evident that in general, popularity is not strongly correlated with budget (0.466139). We want to see whether this correlation varies when going from one genre to another in the top 6. Our team has to be sure that those positive  correlation values carry statistical significance, so that we can justify the pattern between the budget and popularity, specifically that investing more money into film production results in the bigger acclaim of the movie in the future.

```python
list = ["Action", "Adventure", "Drama", "Thriller", "Comedy", "Romance"]
for g in list:
    print(top_6[top_6['genre_1']==g].corr())
    print()
```

We can see that for the top 6 genres, popularity has only a mild correlation with budget. Now, we want to see that this correlation is significant. Let's first have the variables for population correlation coefficient (unknown) and sample correlation coefficient (known).

$\rho =$ population correlation coefficient <br>
$r =$ sample correlation coefficient

If the test says that the correlation coefficient is significantly different from 0, we will say that the correlation coefficient is significant. That is, our hypotheses would be:

Null Hypothesis: $H_0 : \rho = 0$<br>
Alternative Hypothesis: $H_a : \rho \neq 0$<br>

In this hypothesis test, we will use a significance level of $\alpha = 0.05$. If the p-value is less than the significance level ($\alpha = 0.05$), we will reject the null hypothesis and conclude that there is sufficient evidence that there is a significant linear relationship between budget and popularity. Else, we will not reject the null hypothesis and conclude that there is insufficient evidence that there is a significant linear relationship between budget and popularity because the correlation coefficient is not significantly different from 0.

Note that the formula of the test statistic for correlation coefficient is $t = \frac{r\sqrt{n-2}}{\sqrt{1-r^2}}$, where $n$ is the number of samples and $r$ is the correlation coefficient.

2699 out 3491 movies belong to top 6 genres. We used $n = 2699$ data points to calculate the budget-popularity coefficient, which is 0.466139.

```python
n = 2699
r = 0.466139
t = r * (n-2)**(1/2) / (1-r**2)**(1/2)
print("Test statistic t:", t)
```

Both the test statistic $t$ and the correlation coefficient $r$ are positive.

```python
r, p = scipy.stats.pearsonr(top_6["budget"], top_6["popularity"])
print("Pearson's correlation coefficient r:", r)
```

We can confidece-check our correlation coefficient we made above by seeing it is equal to Pearson's correlation printed in the previous cell.

```python
print("p-value from Pearson's correlation:", p)
```

We can see that the p-value from Pearson's correlation is way below 0.05, hence we reject the null hypothesis and conclude that there is a sufficient evidence that there is a significant linear relationship between budget and popularity.

For more information, we also added alternatives to Pearson's r (Spearman's rho and Kendall's tau), all of which correlate with the final conclusion, below.

```python
spearman_rho, spearman_p = scipy.stats.spearmanr(top_6["budget"], top_6["popularity"]) # Spearman's rho
kendall_tau, kendall_p = scipy.stats.kendalltau(top_6["budget"], top_6["popularity"]) # Kendall's tau

print("The value of Spearman's rho is:", spearman_rho)
print("The p-value with the calculation of Spearman's rho is:", spearman_p)

print()
print("The value of Kendall's tau is:", kendall_tau)
print("The p-value with the calculation of Kendall's tau is:", kendall_p)
```

We can see that the p-value for the alternatives are also less than the alpha value 0.05. Ultimately, the correlations are proved to be statistically significant.

## Limitations
The limitations of our study arise from either the orginal raw dataset or our various assumptions about the results from analyses.

- Since our dataset includes around 3,600 movies most of which are popular in society to a certain degree (predominantly in the US market), it might seem like our data is a limited representation of the movies in the market. In other words, there is a possibility of our outcomes and analysis being biased in terms of generating genre trends only from a small sample of movies. There is a chance that the datasets we have used are an anomalous sample taken from a big population of variables. 
- It is important to note that our dataset only includes movies up to year 2016. Therefore, there is a 4-year gap of movies released between 2016 and 2020 - so we don't examine the genres trending during the last four years. 
- We are heavily relying on the dataset’s accurately distributing which movie belongs to which combination of genres; hence, we have to recognize the potential imprecision of results if the genres were incorrectly allocated to a certain movie.
- Our analysis of financial performance as well as budget and revenue calculations of different movie genres are not precise because we do not account for the time value of money (our time period is too wide and external factors like inflation and economy events impacted the value of money). As a result, our assumptions about financial success might be biased and advantageous for the genres of movies released most recently.
- The project analyzes 18 movie genres; it is highly likely that the genres like Foregin, War, History, and Western were underrepresented since only small portions of movies affiliated with these genres. Thus, a big difference in the number of movies under each genre could have brought differences in the final results like correlation coefficients or r^2 score from the linear regression analyses.
- Lastly, it is merely impossible to be able to examine all the movies (even just one country market) because of the immense sizes of the cinema industry. Although we hope that our dataset includes the majority of the most influential movies that has formed genre trends during certain years, there is still a chance that our data analytics had led to erroneous conclusions. If the project's conclusions will be used by someone trying to  predict how a genre of a movie influences its success in the future,  our relatively narrow range of movies might result in harmful consequences for that person. For example,  the user could be misinformed and later producing or participating a movie not reach high popularity and a large box office.
